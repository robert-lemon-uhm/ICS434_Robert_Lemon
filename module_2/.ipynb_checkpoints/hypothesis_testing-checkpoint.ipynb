### Example

* Suppose we have two samples that represent biki ride durations for two distinct neighborhoods `a` and `b`

1. Instead of working with actual data, sample the data from a gaussian the same mean ($\mu=20$) the same same standard deviation ($\sigma = 4$)
    * We will sample 15 bike rides for each of regions `a` and `b`
    

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy import stats
```

```python
np.random.seed(12345) # FOR TESTING, REMOVE LATER

# Set parameters
mean = 20
std_dev = 4
sample_size = 15

# Sample the data
a = np.random.normal(mean, std_dev, sample_size)
b = np.random.normal(mean, std_dev, sample_size)
a,b
```

### Example - cont'd

2. Compute and compare the means of datasets `a` `b`
  * Based on the observed difference between the means, do you believe there is a statistically significant difference between samples `a` and `b`

```python
# Compute and compare means of a and b
a_mean = a.mean()
b_mean = b.mean()
print("The mean of group a is ", a_mean)
print("The mean of group b is ", b_mean)
```

By simply looking at the means of the two datasets, I would not think that there is a statistically significant difference between the two. They are also within 1 standard deviation of each other.

### Example - cont'd

3. draw the two KDEs reprenting the datasets `a` and `b`
  * You should have a figure that appears "approximately" as in the figure below
    * The X- and Y-axis values have been intentionally omitted
    
![](https://www.dropbox.com/s/0ab4d92lkcn89jv/a_b_kde_exmaple.png?dl=1)    

```python
# Generate KDEs for datasets
a_kde = sp.stats.gaussian_kde(a)
b_kde = sp.stats.gaussian_kde(b)

x_values = np.arange(10,30, 0.25)

a_densities = a_kde.evaluate(x_values)
b_densities = b_kde.evaluate(x_values)

# Graph KDEs
plt.figure(figsize=(16,8))
plt.plot(x_values, a_densities)
plt.plot(x_values, b_densities)
plt.xlabel("X")
plt.ylabel("Density")
plt.title("KDE Graphs of A and B")
```

### Example - cont'd

4. Visually, does dataset *a* differ significantly from dataset `b`

Visually, dataset a does not appear to significantly differ from dataset b. Their mean seems to be around the same value and the shape and height of their KDE curves are similar.
